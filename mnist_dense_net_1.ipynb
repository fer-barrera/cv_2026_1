# DEEP LEARNING 1

Importa las librerías necesarias para cargar MNIST, construir una red neuronal con Keras, procesar datos y manejar visualización y archivos en Google Colab.

```python
from keras import layers
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
```

Carga el conjunto de datos MNIST dividiéndolo en imágenes y etiquetas para entrenamiento y prueba.

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

Obtiene y muestra las dimensiones de los datos MNIST, incluyendo cantidad de imágenes y tamaño de cada imagen.

```python
num_train_images = X_train.shape[0]
num_test_images = X_test.shape[0]
image_height = X_train.shape[1]
image_width = X_train.shape[2]
print("X_train.shape: " + str(X_train.shape))
print("X_test.shape: " + str(X_test.shape))
print("y_train.shape: " + str(y_train.shape))
print("y_test.shape: " + str(y_test.shape))
print("Train samples: " + str(num_train_images))
print("Test samples: " + str(num_test_images))
print("Image height: " + str(image_height))
print("Image width: " + str(image_width))
```

Reestructura cada imagen MNIST en un vector plano y convierte los datos a tipo flotante de 32 bits.

```python
X_train = X_train.reshape(num_train_images, image_height * image_width).astype('float32')
X_test = X_test.reshape(num_test_images, image_height * image_width).astype('float32')
```

```python
print("X_train.shape: " + str(X_train.shape))
print("X_test.shape: " + str(X_test.shape))
print("y_train.shape: " + str(y_train.shape))
print("y_test.shape: " + str(y_test.shape))
```

Normaliza los valores de los píxeles escalándolos al rango [0, 1].

```python
X_train = X_train / 255
X_test = X_test / 255
```

Convierte las etiquetas numéricas en vectores one-hot para entrenamiento del modelo.

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

Imprime la forma y ejemplos de las etiquetas one-hot

```python
print("categorical y_train shape: " + str(y_train.shape))
print("categorical y_train sample 5 value: " + str(y_train[5]))
print("categorical y_train sample 2 value: " + str(y_train[2]))
num_classes = y_test.shape[1]
print(num_classes)
```

Construye una red neuronal secuencial con dos capas densas ocultas, dropout y una capa final softmax para clasificar dígitos del MNIST.

```python
model = Sequential()
model.add(Dense(500, input_dim=image_height*image_width, activation='relu'))
model.add(Dense(20, input_dim=image_height*image_width, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(10,  activation='softmax'))
```

Muestra un resumen estructurado de las capas, parámetros y arquitectura del modelo.

```python
print(model.summary())
```

Compila y entrena el modelo durante 50 épocas usando Adam y entropía cruzada, registrando la precisión y validación.

```python
EPOCHS = 50
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=EPOCHS, batch_size=512, verbose=0)
```

Genera gráficos comparando la pérdida y exactitud de entrenamiento y validación a lo largo de todas las épocas.

```python
def plot(h):
    LOSS = 0; ACCURACY = 1
    training = np.zeros((2,EPOCHS)); testing = np.zeros((2,EPOCHS))
    training[LOSS] = h.history['loss']
    testing[LOSS] = h.history['val_loss']    # validation loss
    training[ACCURACY] = h.history['accuracy']
    testing[ACCURACY] = h.history['val_accuracy']  # validation accuracy

    epochs = range(1,EPOCHS+1)
    fig, axs = plt.subplots(1,2, figsize=(15,5))
    for i, label in zip((LOSS, ACCURACY),('loss', 'accuracy')):
        axs[i].plot(epochs, training[i], 'b-', label='Training ' + label)
        axs[i].plot(epochs, testing[i], 'y-', label='Test ' + label)
        axs[i].set_title('Training and test ' + label)
        axs[i].set_xlabel('Epochs')
        axs[i].set_ylabel(label)
        axs[i].legend()
    plt.show()
```

```python
plot(history)
```
