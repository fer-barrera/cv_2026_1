```python
# Importar las librerías necesarias
import cv2
```

## 1. Lectura de la imagen

```python
img = cv2.imread('.\images\circles.png')
assert img is not None, "la imagen no pudo ser leida, verifique la ruta"
```

## 2. Atributos de una imagen

### Shape (rows, columns, and channels)

```python
print( img.shape )
```

### Número de filas y de columnas

```python
print(f'Alto: {img.shape[0]} píxeles')
print(f'Ancho: {img.shape[1]} píxeles')
```

### Número de canales

```python
# verificar la dimensión de la imagen
if len(img.shape) == 2:
    num_canales = 1
else:
    num_canales = img.shape[2]
    print(f'Número de canales: {num_canales}')
```

### Tipo de dato

```python
print(f'Tipo de dato: {img.dtype}')
```

### Número total de píxeles

```python
print(f'Número total de píxeles: {img.size}')
```

### Tamaño en bytes

```python
print(f'Tamaño de la imagen en bytes: {img.nbytes}')
```
